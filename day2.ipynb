```python
def calculate_carbon_footprint(energy_consumption,emission_factor):
    """
    this function calaculate the carbon footprint based on energy consumption
    (in kilowatt-hours) and the emission factor (in kg CO2 per kWh).
    """
    return energy_consumption * emission_factor
```

```python
energy_consumption = 100
emission_factor = 0.475
carbon_footprint = calculate_carbon_footprint(energy_consumption,emission_factor)
print(f"Carbon Footprint: {carbon_footprint}")
```

```python
import pandas as pd
import numpy as np
data = {
    "Energy Sources" : ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "NEclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 32000],
    "Cost (Million $)": [200,400, np.nan, 150, 250, np.nan]
}
energy_df = pd.DataFrame(data)
print("Original Energy data with missing values:")
print(energy_df)
```

```python
energy_df.head()
```

```python
energy_df.isnull().sum()
```

```python
energy_df.isnull().sum().sum()
```

```python
energy_df.tail()
```

```python
cleaned_df = energy_df.dropna()
print("\nData after removing rows with missing values:")
print(cleaned_df)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#impute missing values with mean
energy_df["Energy Consumption (MWh)"].fillna(energy_df["Energy Consumption (MWh)"].mean(),inplace=True)
energy_df["Cost (Million $)"].fillna(energy_df["Cost (Million $)"].mean(),inplace=True)
print("\nData after imputing missing values with mean:")
print(energy_df)
```

```python
#forward fill and backwordfill
forward_filled_df = energy_df.fillna(method="ffill")
print("\nData after forward filling:")
print(forward_filled_df)
```

```python
backward_filled_df = energy_df.fillna(method="bfill")
print("\nData after backword filling:")
print(backward_filled_df)
```

```python
energy_df["Missing Consumption"] = energy_df["Energy Consumption (MWh)"].isna().astype(int)
print("\nData with missing values flagged:")
print(energy_df)
```

```python
#normalizing(min-max scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]]
)
print("\nData after normalization (Min-Max Scaling):")
print(energy_df)
```

```python
#standarization(z-score scaling)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]] = scaler.fit_transform(
    energy_df[["Energy Consumption (MWh)", "Cost (Million $)"]]
)
print("\nData after standardization (z-score scaling):")
print(energy_df)
```

```python
energy_encoded_df = pd.get_dummies(energy_df, columns=["Energy Sources"])
print("\nData after one-hot encoding categorical variables:")
print(energy_encoded_df)
```

```python
#feature engineering
energy_encoded_df["Consumption per $Million"] = energy_encoded_df["Energy Consumption (MWh)"] / energy_encoded_df["Cost (Million $)"] 
print("\nData with new feature (Consumption per Million$):")
print(energy_encoded_df)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
# Creating a DataFrame for pair plot
data={
    "Solar": [1200, 1500, 1300],
    "Wind": [3400, 3600, 3200],
    "Hydropower": [2900, 3100, 2800],
    "Biomass": [2500, 2700, 2400]
}
df = pd.DataFrame(data)
#f Create a pair plot

sns.pairplot(df)


plt.show()
```

```python
#heapmap showing differnt type of data
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidth=0.5)
plt.title('Correlation Between Energy Sources')
plt.show()
```

```python
# box plot use to check outlires
sns.boxplot(data=df)
plt.title('Distribution of Energy Consumption by Source')
plt.show()
```

```python
#ki kitna spread hua hi data
sns.violinplot(data=df)
plt.title('Violin plot of energy consumption distribution')
plt.show()
```

```python
# use to show best straight line or best fit line to show the trand of data
energy_values = [100,200,300,400,500]
carbon_emissions = [10,20,30,40,50]
df_reg = pd.DataFrame({
    'Energy Consumption': energy_values,
    'Carbon Emission': carbon_emissions
})
sns.regplot(x='Energy Consumption', y='Carbon Emission', data = df_reg)
plt.title('Energy Consumption vs Carbon Emissions')
plt.xlabel('Energy Consumption')
plt.ylabel('Carbon Emissions')
plt.show()
```

```python
#facet grid : plotting energy consumption over time(hue = is for filter)
months = ['January','Februry','March','April','May','June']
df_facet = pd.DataFrame({
    'Month' : months *3,
    'Energy Consumption':[1200, 1300, 1100, 1500, 1400, 1600] * 3,
    'Region' : ['North','South','East'] * 6
})
g = sns.FacetGrid(df_facet, col='Region', hue='Region', height=4,aspect = 1)
g.map(sns.lineplot, 'Month', 'Energy Consumption')
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Energy Consumption Over Time by Region')
plt.show()
```
