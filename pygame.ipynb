```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600

# Colors
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

# Ball properties
BALL_RADIUS = 20
ball_x = WIDTH // 2
ball_y = BALL_RADIUS
ball_speed_y = 0

# Physics properties
gravity = 0.5
elasticity = 0.8

# Clock for controlling frame rate
clock = pygame.time.Clock()

# Create the screen
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Elastic Ball Bounce")

# Main loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Physics update
    ball_speed_y += gravity
    ball_y += ball_speed_y

    # Bounce on hitting the ground
    if ball_y + BALL_RADIUS >= HEIGHT:
        ball_y = HEIGHT - BALL_RADIUS
        ball_speed_y = -ball_speed_y * elasticity

    # Clear the screen
    screen.fill(WHITE)

    # Draw the ball
    pygame.draw.circle(screen, BLUE, (ball_x, int(ball_y)), BALL_RADIUS)

    # Update the display
    pygame.display.flip()

    # Cap the frame rate
    clock.tick(60)

# Quit Pygame
pygame.quit()
sys.exit()
```
